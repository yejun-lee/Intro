### 04 - 1 데이터 시각화가 필요한 이유 : 앤스콤 4분할 그래프 (81p)

```python
anscombe = sns.load_dataset('anscombe')
anscombe
```

#### 데이터 1 시각화

```python
data_1 = anscombe[anscombe['dataset']=='I']
data_1 = data_1.sort_values(by='x')
data_1
```

```python
plt.figure(facecolor='white')

plt.plot(data_1.x, data_1.y, marker='o')
plt.show()
```

#### 전체 데이터 시각화

```python
data_2 = anscombe[anscombe['dataset']=='II']
data_3 = anscombe[anscombe['dataset']=='III']
data_4 = anscombe[anscombe['dataset']=='IV']
```

```python
plt.figure(figsize=(8, 6), facecolor='white')
plt.suptitle('Anscombe Data')

plt.subplot(2, 2, 1)
plt.plot(data_1.x, data_1.y, 'o')
plt.title('< data_1 >')
plt.subplot(2, 2, 2)
plt.plot(data_2.x, data_2.y, 'o')
plt.title('< data_2 >')
plt.subplot(2, 2, 3)
plt.plot(data_3.x, data_3.y, 'o')
plt.title('< data_3 >')
plt.subplot(2, 2, 4)
plt.plot(data_4.x, data_4.y, 'o')
plt.title('< data_4 >')

plt.tight_layout()
plt.show()
```

### 네 가지 데이터의 평균, 분산, 상관관계, 회귀선의 수치가 같아도 그래프의 형태는 다를 수 있다.

```python
print(data_1.describe())
print(data_2.describe())
print(data_3.describe())
print(data_4.describe())
```

```python
print(data_1.corr())
print(data_2.corr())
print(data_3.corr())
print(data_4.corr())
```

### 04 - 2 matplotlib 라이브러리 자유자재로 사용하기

```python
tips = sns.load_dataset('tips')
tips.head()
```

#### 히스토그램 : 변수 1개

```python
fig = plt.figure(facecolor='white')

plt.hist(tips['total_bill'], bins=20)
plt.title('Histogram of Total Bill')
plt.xlabel('Frequency')
plt.ylabel('Total Bill')

plt.show()
```

#### 산점도 그래프 : 변수 2개

```python
plt.figure(facecolor='white')

plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Scatterplot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

#### Box Plot : 이산형 변수와 연속형 변수를 함께 사용

```python
plt.figure(facecolor='white')

plt.boxplot([tips[tips['sex']=='Female']['tip'], 
            tips[tips['sex']=='Male']['tip']], 
           labels=['Female', 'Male'])

plt.title('Boxplot of Tips by Sex')
plt.xlabel('Sex')
plt.ylabel('Tip')

plt.show()
```

#### 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프

```python
tips['sex_color'] = tips['sex'].map({'Female':0, 'Male':1})
tips
```

```python
plt.figure(facecolor='white')

# s : Size(점의 크기), c : Color(점의 색상)
plt.scatter(x=tips['total_bill'], 
           y=tips['tip'], s=tips['size']*10, c=tips['sex_color'], alpha=0.5)

plt.title('Total Bill vs Tip Colored by Sex and Sized by Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
