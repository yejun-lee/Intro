## dictionary
- 키와 값을 갖는 데이터 구조
- 키는 내부적으로 hash값으로 저장 : **동일한 키**가 중복해서 있을 수 없음
- 순서를 따지지 않음. 즉, 인덱스가 없음
- **리스트는** 원하는 원소가 몇 번째 인덱스에 있는지 아는 것은 오래 걸릴 수도 있지만 내가 찾고자 하는 원소가 몇 번째 인덱스인지 안다면 접근하는 데 걸리는 시간은 거의 0에 가깝다.
- **dictionanry**는 키만 알면 원하는 값을 바로 가져올 수 있다.

````python
a = {'Korea': 'Seoul', 
     'Canada': 'Ottawa',
     'USA': 'Washington D.C'}

print(a)
type(a)
print(a['Korea'])
````

- **항목 추가 및 변경**
  - 기존에 키가 존재하면, 새로운 값으로 **업데이트**
  - 존재하지 않으면 새로운 키, 값 **생성**

````python
a = {'Korea': 'Seoul', 
     'Canada': 'Ottawa',
     'USA': 'Washington D.C'}

a['Japan'] = "Tokyo"
a['China'] = 'Beijing'

print(a)
````

- **update()**
  - 두 dictionary를 병합함
  - 겹치는 키가 있다면 parameter로 전달되는 키 값이 overwrite된다.

````python
a = {'a': 1, 'b': 2, 'c': 3}
b = {'a': 2, 'd': 4, 'e': 5}

a.update(b)

print(a)
````

- **key삭제**
  - del 키워드 사용
  - pop 함수 이용

````python
a = {'a': 2, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

del a['b']
a.pop('d')

print(a)
````

- **clear()**
  - dictionary의 모든 값을 초기화

````python
a = {'a': 2, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
print(a)

a.clear()
print(a)
````

  - 리스트의 경우에는 리스트의 길이가 길어질 수록 성능저하가 초래됨

## set
- dictionary에서 key만 활용하는 데이터 구조로 이해
- 수학에서의 집합과 동일한 개념
- set은 안의 원소의 순서는 보장하지 않음
- 즉, list처럼 인덱스로 접근할 수가 없음

````python
a = {1, 1, 2, 3, 3, 4, 1, 5}
print(a)

type(a)
````

- set()으로 집합으로 변환

````python
a = [1, 1, 2, 3, 3, 4, 1, 5]
print(a)

b = set(a)
print(b)
type(b)
````

````python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.union(b))   # 합집합 (a | b)
print(a.intersection(b))   # 교집합 (a & b)
print(a.difference(b))   # 차집합 (a - b)

print(a.issubset(b))   # 부분집합 여부
````

## 함수의 정의
- 정의 시 최초에 def 키워드 사용

    - 함수를 호출한 코드 (caller)로 함수가 해당 기능을 수행하고 완료된 값(output)을 전달하기 위해 return 키워드 사용

````python
def add(a, b):
    return a + b
````

````python
print(add(20, 400))
````

## parameter(argument) (인자)
- 함수에 전달되는 입력(input)
- 입력이 필요하지 않을 수도 있고, 여러 개의 입력이 필요할 수도 있음.
- 파라미터로 int, string, float, bool, list, dict 등 어떤 파이썬 객체도 전달 가능
- 심지어, 함수도 함수의 파라미터로 전달 가능
- python의 경우, 타입 명시가 없기 때문에 함수 생성 시 의도된 파라미터의 타입에 맞게 입력을 전달하는 것이 중요
- 또한 파라미터를 전달할 때, 정의된 순서에 따라 값을 전달하는 것이 중요

### Default parameter (기본 인자)
- 함수의 파라미터에 기본 값 지정 가능
- 파라미터를 명시하지 않을 경우, 지정된 기본 값으로 대체

````python
def add(x, y=10, z=5):
    return x + y + z

print(add(10, 1, 2))
print(add(20))
````

#### Default parameter 사용 시 주의점
- 디폴트 파라미터 뒤에 일반 파라미터가 위치할 수 없음
- e.g) 올바른 예
> def test(a, b, c=1)

    > def test(a, b=1, c=2)

    > def test(a=1, b=1, c=3)

- e.g) 올바르지 않은 예
> def test(a, b=1, c)

    > def test(a=1, b, c)

    > def test(a=1, b=1, c)

### keyword parameter (키워드 파라미터)
- 파이썬의 경우, 파라미터에 값을 전달할 때 파라미터의 이름을 명시하여 전달 가능
- 파라미터 이름을 사용하지 않을 경우, 기본적으로 순서에 맞게 전달
```python
def test(x, y, z):
    return x + y + z
test(x=3, z=2, y=5)
```

### return
- 기본적으로 함수의 종료를 명시
    - return 옆에 값이나 수식이 있다면 해당 값을 호출자(caller)에게 반환(전달)
    - return 만 존재하면 None 반환
    - return이 없는 경우, 기본적으로 함수 코드 블록이 종료되면 종료로 간주. 이때도 None 반환

#### multiple return (복수 값 반환)
- tuple 반환을 하여 복수 개의 값 리턴 가능

````python
def add_mul(x, y):
    s = x + y
    m = x * y
    return s, m

c = add_mul(20, 3)
print(type(c))
print(c)

# tuple unpacking
a, b = add_mul(15, 23)
print(type(a))
print(a, b)
````

### variable scope (변수의 범위)
- 변수가 참조 가능한 코드 상의 범위를 명시
- 함수 내 변수는 자신이 속한 코드 블록이 종료되면 소멸됨
- 이렇게 특정 코드 블록에서 선언된 변수를 **지역변수(local variable)**이라고 함
- 반대로 가장 상단에서 정의되어 프로그램 종료 전까지 유지되는 변수를 **전역변수(global variable)**이라고 함
- 같은 이름의 지역변수와 전역변수가 존재할 경우, 지역변수의 우선순위가 더 높음

````python
num1, num2 = 10, 30   # 전역 변수

def test(num1, num2):   # 지역 변수
    print(num1, num2)
    return num1 + num2

test(20, 40)

print(num1, num2)
````

### variable length argument (가변길이 인자)
- 전달되는 파라미터의 개수가 고정적이지 않은 경우 사용
- e.g)
    - print 함수
    - format 함수
    > ***args**, ****kwargs**
    
    > ***args** : 파라미터를 튜플의 형태로 전달
    
    > ****kwargs** : 파라미터를 dict 형태로 전달(네임드 파라미터)

````python
# def print(가변길이 파라미터):

print()
print(1)
print(1, 2)
print(1, 2, 3)
print(1, 2, 3, 4)
print(1, 2, 3, 4, 5)
````

````python
def test(*x):
    print(type(x))

test(10)
test(10, 20)
test(10, 20, 30)
````

````python
def test(*args):   # 가변길이 함수를 사용할 때는 관례적으로 *args라 표기해야 한다.
    for item in args:
        print(item)

test(10, 20, 30, 49)
````

### keyword parameter (키워드 파라미터)
- **가 붙은 경우에는 키워드 파라미터로 인식
- 즉 함수 호출 시, 파라미터의 이름과 값을 함께 전달 가능

### filter, map, reduce
- lambda가 유용하게 사용되는 3가지 대표적 함수
- 함수형 프로그래밍의 기본 요소이기도 함
- filter : 특정 조건을 만족하는 요소만 남기고 필터링
- map : 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환
- reduce : 차례대로 앞 2개의 원소를 가지고 연산. 연산의 결과가 또 다음 연산의 입력으로 진행됨. 따라서 마지막까지 진행되면 최종 출력은 한 개의 값만 남게 됨
```python
filter(함수, 리스트)  : 참이 되면 살아남고 거짓이면 날라감
map(함수, 리스트) : 원래의 리스트를 함수를 거친 새로운 리스트로 반환
reduce(함수, 리스트) : 마지막 한 개의 값이 남을 때까지 앞에서 부터 차례대로 함수를 거침
```

````python
## filter 함수
nums = [1, 2, 3, 6, 8, 9]

print(filter(lambda x:x%2==0, nums))
print(list(filter(lambda x:x%2==0, nums)))
````

````python
## map 함수
nums = [1, 2, 3, 6, 8, 9]

print(map(lambda x:x**2, nums))
print(list(map(lambda x:x**2, nums)))
````

````python
from functools import reduce

nums = [1, 2, 4, 5, 6, 8, 10]

print(reduce(lambda x,y:x+y, nums))
````

## 함수 연습문제
1. 주어진 숫자 리스트의 평균을 구하는 함수를 출력하시오
2. 해당 숫자가 소수인지 아닌지 판별하시오
3. 2부터 해당 숫자 사이에 소수가 몇 개인지 출력하는 함수를 구하시오

1. 주어진 숫자 리스트의 평균을 구하는 함수를 출력하시오

````python
def average(x):
    _sum = 0
    for i in x:
        _sum += i
    return _sum

nums = [1, 2, 4, 5, 6, 8, 10]

avgs = average(nums) / len(nums)
print(avgs)
````

2. 해당 숫자가 소수인지 아닌지 판별하시오

````python
def prime_num(x):
    a = 0
    for i in range(1, x+1):
        if x % i == 0:
            a += 1
        else:
            continue
    
    if a == 2:
        return True
    elif a > 2:
        return False
    elif a == 1:
        return None

print(prime_num(23))
print(prime_num(27))
print(prime_num(1))
print(prime_num(533))
````

3. 2부터 해당 숫자 사이에 소수가 몇 개인지 출력하는 함수를 구하시오

````python
def num_of_prime(x):
    a = 0
    for i in range(2, x+1):
        if prime_num(i) == True:
            a += 1
    return a

print(num_of_prime(10))
print(num_of_prime(100))
print(num_of_prime(150))
print(num_of_prime(5000))
````

## `__init__`(self)
- 생성자, 클래스 인스턴스가 생성될 때 호출됨
- self인자는 항상 첫 번째에 오며 자기 자신을 가리킴
- 이름이 꼭 self일 필요는 없지만, 관례적으로 self로 사용
- 생성자에서는 해당 클래스가 다루는 데이터를 정의
    - 이 데이터를 멤버 변수(member variable) 또는 속성(attribute)라고 함

````python
class Person:
    def __init__(self):
        print(self, 'is generated')
        self.name = 'Kate'
        self.age = 10

p1 = Person()
p2 = Person()

p1.name = 'Aaron'
p1.age = 20

print(p1.name, p1.age)
print(p2.name, p2.age)
````

- 이름과 나이를 동적으로 설정하고 싶을 때

````python
class Person:
    def __init__(self, name, age):   # Default Parameter 설정 가능
        print(self, 'is generated')
        self.name = name
        self.age = age

p1 = Person('Bob', 30)
p2 = Person('Kate', 10)

print(p1.name, p1.age)
print(p2.name, p2.age)
````

### self
- 파이썬의 method는 항상 첫 번째 인자로 self를 전달
- self는 현재 해당 메쏘드가 호출되는 객체 자신을 가리킴
- C++/C#, Java의 this에 해당
- 이름이 self일 필요는 없으나, 위치는 항상 맨 처음의 parameter이며 관례적으로 self로 사용

````python
class Person:
    def __init__(self, name, age):
        print(self)  # self는 객체 자체를 나타내는 것
        self.name = name
        self.age = age
    def sleep(self):
        print(self.name, '은 잠을 잡니다.')

a = Person('Aaron', 20)
b = Person('Bob', 30)

print(a)
print(b)

a.sleep()
b.sleep()
````

## method 정의
- 멤버함수라고도 하며, 해당 클래스의 object에서만 호출 가능
- method는 객체 레벨에서 호출되며, 해당 객체의 속성에 대한 연산을 행함
- {obj}.{method}() 형태로 호출됨

````python
class Counter:
    def __init__(self):
        self.num = 0

    def increment(self):
        self.num += 1

    def reset(self):
        self.num = 0

    def print_current_value(self):
        print('현재 값은: ', self.num)

c1 = Counter()
c1.print_current_value()
c1.increment()
c1.increment()
c1.print_current_value()

c2 = Counter()
c2.print_current_value()
````
